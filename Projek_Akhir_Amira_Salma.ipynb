- Pertanyaan 1: Bagaimana kecepatan angin memengaruhi rata-rata jumlah sepeda yang disewakan?
- Pertanyaan 2: Apakah terdapat tren peningkatan atau penurunan dalam jumlah sepeda yang disewakan dari bulan ke bulan pada tahun 2011?

## Import Semua Packages/Library yang Digunakan

```python
# Import Packages/Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
hour = pd.read_csv('https://raw.githubusercontent.com/amirasalmaa/Bike-Sharing-Dataset/main/hour.csv')
hour.head()
```

### Assessing Data

```python
hour.info()
hour.isna().sum()
print("Jumlah duplikasi: ", hour.duplicated().sum())
```

### Cleaning Data

Tahapan Cleaning Data tidak diperlukan karena dataset yang diberikan sudah tidak terdapat *missing value* maupun data duplikasi.

## Exploratory Data Analysis (EDA)

### Rangkuman Parameter Statistik Data hour

```python
hour.describe(include="all")
```

### Eksplorasi

```python
# Eksplorasi untuk pertanyaan 1
# Menghitung rata-rata jumlah sepeda yang disewakan untuk setiap kecepatan angin
avgwindspeed = hour.groupby('windspeed')['cnt'].mean()

# Membuat Data Frame
avgwindspeed_df = pd.DataFrame({
    'Kecepatan Angin': avgwindspeed.index,
    'Rata-rata Jumlah Sepeda Disewakan': avgwindspeed.values
})

# Menampilkan Data Frame
print(avgwindspeed_df)
```

```python
# Eksplorasi untuk pertanyaan 2
# Membuat kolom "Bulan" dan "Tahun" dari kolom "Tanggal" yang sudah dikonversi menjadi tipe data datetime
hour['Tanggal'] = pd.to_datetime(hour['dteday'])
hour['Bulan'] = hour['Tanggal'].dt.month
hour['Tahun'] = hour['Tanggal'].dt.year

# Memfilter data hanya untuk tahun 2011
data2011 = hour[hour['Tahun'] == 2011]

# Menghitung jumlah sepeda yang disewakan setiap bulan pada tahun 2011
tren_penyewaan_bulanan_2011 = data2011.groupby('Bulan')['cnt'].sum()

# Membuat Data Frame
tren = pd.DataFrame({
    'Bulan': tren_penyewaan_bulanan_2011.index,
    'Jumlah Sepeda Disewakan' : tren_penyewaan_bulanan_2011.values
})

# Menampilkan Data Frame
print(tren)
```

```python
# Eksplorasi untuk pertanyaan 3
# Menghitung jumlah sepeda yang disewakan setiap jenis pengguna
TotalTerdaftar = hour['registered'].sum()
TotalBiasa = hour['casual'].sum()

# Membuat Data Frame
Perbandingan = pd.DataFrame({
    'Jenis Pengguna' : ['Terdaftar', 'Biasa'],
    'Jumlah Sepeda Disewakan' : [TotalTerdaftar, TotalBiasa]
})

# Menampilkan Data Frame
print(Perbandingan)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisai Pertanyaan 1
plt.figure(figsize=(10, 5))
plt.plot(avgwindspeed_df['Kecepatan Angin'], avgwindspeed_df['Rata-rata Jumlah Sepeda Disewakan'], marker='o', linewidth=2)
plt.title('Rata-rata Jumlah Sepeda Disewakan Berdasarkan Kecepatan Angin')
plt.xlabel('Kecepatan Angin')
plt.ylabel('Rata-rata Jumlah Sepeda Disewakan')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan visualisasi di atas, dapat terlihat bahwa rata-rata jumlah sepeda disewakan terbanyak terjadi pada saat kecepatan angin tertinggi.

### Pertanyaan 2:

```python
# Visualisasi Pertanyaan 2
plt.figure(figsize=(10, 5))
plt.plot(tren['Bulan'], tren['Jumlah Sepeda Disewakan'], marker='o', linewidth=2)
plt.title('Tren Jumlah Penyewaan Sepeda di Tahun 2011')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Sepeda Disewakan')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan visualisai di atas, dapat terlihat bahwa terjadi tren penaikan pada bulan 1 sampai dengan bulan 6 dan terjadi tren penurunan pada bulan 7 sampai dengan bulan 12.

### Pertanyaan 3:

```python
plt.figure(figsize=(10, 5))
plt.bar(Perbandingan['Jenis Pengguna'], Perbandingan['Jumlah Sepeda Disewakan'], color=['blue', 'orange'])
plt.title('Perbandingan Jumlah Sepeda Disewakan Berdasarkan Jenis Pengguna')
plt.xlabel('Jenis Pengguna')
plt.ylabel('Jumlah Sepeda Disewakan')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan visualisai di atas, dapat terlihat bahwa jumlah sepeda yang disewakan lebih banyak oleh jenis pengguna yang sudah terdaftar.
